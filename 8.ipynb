### Univariate Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("dataset/train.csv")
```

```python
df.sample(5)
```

## 1. Categorical data : 

Freq. count for each categories in a feature/class

#### a. Countplot

```python
sns.countplot(x="Survived",data=df,color="blue")
df["Survived"].value_counts()
```

```python
sns.countplot(x="Pclass",data=df)
```

#### b. PieChart

```python
df["Survived"].value_counts().plot(kind="pie",autopct="%1.1f%%")
```

### 2. Numerical Data

#### a. Histogram : 

To check distribution of numeric data

```python
plt.hist(df["Age"],bins=10)
```

#### b. Distplot

Gives the pdf, prob.of having an event.

```python
sns.displot(df["Age"])
```

#### c. Boxplot

```python
sns.boxplot(df["Fare"])
```

```python
print(df['Age'].mean())
print(df['Age'].max())
print(df['Age'].skew())
```
