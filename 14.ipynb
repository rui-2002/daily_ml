```python
pd.get_dummies(df,columns=["fuel","owner"],dtype=int)
```

#### 2. K-1 OneHotEncoding

```python
pd.get_dummies(df,columns=["fuel","owner"],drop_first=True,dtype=int)
```

we don't use pandas while using OneHotEncoding as padas don't store how the columns are placed ,result can be changed at retry. So we using scikit-learn class

#### 3. OneHotEncoding using Scikit-learn

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,0:4],df.iloc[:,:-1],test_size=0.2,random_state=44)
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()
X_train_new=ohe.fit_transform(X_train[["fuel","owner"]]).toarray()
```

```python
X_test_new=ohe.transform(X_test[["fuel"]])
```
