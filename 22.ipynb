```python
# Numerical to categorical : 2 Techniques :
# 1. Discretization (binning)
# 2. Binnarization
```

Discretization is the process of transforming continious variables into discrete variable by creating a set of contiguous intervals that span the range of the variable's values. Discretization is also called binning, whre bin is an alternative name for interval.



Why use Discretization :
1. To handle Outliers
2. To improve the values spread
