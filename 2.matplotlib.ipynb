## Matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
# %matplotlib inline # this line is used to when we don't want to write plt.show()
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
x,y
```

```python
# plotting using matplotlib

plt.scatter(x,y,c='g')
plt.xlabel("X axis")
plt.ylabel("Y label")
plt.title("Graph in 2D")
```

```python
## Creating subplots:
plt.subplot(2,2,1) # (no. of rows, no. of columns ,positon where we want to place it)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'b')
plt.subplot(2,2,3)
plt.plot(x,y,'g')
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title("sine wave form")


# plt the points using matplotlib
plt.plot(x,y)
```

```python
## Bar plot :


x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]

plt.subplot(2,2,1)
plt.bar(x,y)
plt.xlabel("X axis")
plt.ylabel("Y label")


plt.subplot(2,2,2)
plt.bar(x,y,color='g')
plt.xlabel("X axis")
plt.ylabel("Y label")
```

## Boxplot using matplolib

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);
```

```python
## pie charts

labels='Python','C++','Ruby','Java'
sizes=[215,130,245,210]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.1,0,0,0) # explode 1st slice (to split the pie chart)

# plot 
plt.pie(sizes,explode=explode,labels=labels,colors=colors,
        autopct='%1.1f%%',shadow=True)

plt.axis('equal')
plt.show()
```

## Seaborn 

```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

#### Correlation with Heatmap

A correlation heatmap uses colored cells, typically in a monochromatic scale , to show a 2D correlation matrix (table) between two discrete dimensions or event types. It is very important in Feature Selection.

```python
df.head()
```

```python
df1=df.select_dtypes(include=["number"])
df1.head()
```

```python
df1.corr()
```

```python
sns.heatmap(df1.corr())
```

### Join Plot

A joint plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplt, a hexbin plot, 2D histogram or a 2D density plot.

##### Univariate Analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df1,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df1,kind='reg')
```

### Pair Plot (Feature >2)

A pair plot is also known as a scatteplot, in which one variagle in the same data row is matched with other variable value, like this : pair plot are just elaobarions on this,showing all variables paired with all other variables.

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue="sex")
```

### Distplot
Dist plot us to check the distribution of the columns feature.

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

### Countplot

```python
sns.countplot(x='sex', data=df)
plt.show()
```

```python
sns.barplot(x='sex', y='total_bill',data=df)
plt.show()
```

## Boxplot :

```python
sns.boxplot(x='sex',y='total_bill',data=df)
```

## Violine Plot :

Violin plot help us to see both the distributors of data in terms of Kernel Density Estimation and the box plotmm
